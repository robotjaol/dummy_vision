## Image Processing

in this section we are going to discuss about image processing techniques be later implemented and used in Object Detection and Tracking

```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# import the image

img = cv2.imread('puppy.jpg')
plt.imshow(img)
```

```python
# now convert to RGB format

img1 = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img1)
```

```python
# now convert to HSV format

img2 = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
plt.imshow(img2)
```

```python
# now convert to HLS format

img3 = cv2.cvtColor(img, cv2.COLOR_BGR2HLS)
plt.imshow(img3)
```
